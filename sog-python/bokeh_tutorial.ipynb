## Bokeh

Bokeh allows to create javascript visualizations using python.


Installation:
```shell
pip install bokeh
```

````python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

output_notebook()
````

Now bokeh is ready to use.

````python


def plot(xs, ys):
    p = figure(title='quadratic function')
    p.line(xs, ys, legend_label='x^2 + 6x + 9')
    p.line(xs, ys + 100, legend_label='x^2 + 6x + 9 + 100')
    p.legend.click_policy = 'hide'
    show(p)
    
    
plot(np.linspace(-10, 10, 100), (np.linspace(-10, 10, 100) + 3) ** 2)
````

This plot in interactive, you can drag it or click on a legend to hide some lines.
