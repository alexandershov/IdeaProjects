````python
import graphviz

def show_graph():
    # DOT is a language used by graphviz to describe graphs
    # Digraph is Directed Graph
    # there is also graphviz.Graph for undirected graph
    dot = graphviz.Digraph()
    dot.node('A', 'A label')
    dot.node('B', 'B label')
    dot.edge('A', 'B', 'AB edge')
    
    # we can have composite nodes
    # <name> is a field name
    dot.node('PersonDesc', '<name> name|<age> age', shape='record')
    dot.node('ActualPerson', '<name> sasa|<age> 38', shape='record')
    
    # edge from the given field to the given field
    dot.edge('PersonDesc:name', 'ActualPerson:name')
    dot.edge('PersonDesc:age', 'ActualPerson:age')
    
    # nodes can have different shapes, here C shape is rectangle
    dot.node('C', 'C label', shape='rect')
    
    # we can create several edges in one call
    # in the edge C -> PersonDesc:name we used a compass, C arrow go into PersonDesc:name from the west
    # n/s/w/e/ne/nw/se/sw are supported
    dot.edges([('A', 'C'), ('C', 'PersonDesc:name:w')])
    
    return dot

show_graph()
````

````python
def visualize_leetcode_maximum_employees_to_be_invited_to_a_meeting(favorites):
    # https://leetcode.com/problems/maximum-employees-to-be-invited-to-a-meeting/
    dot = graphviz.Digraph()
    for i in range(len(favorites)):
        dot.node(str(i))
    for i, fav in enumerate(favorites):
        dot.edge(str(i), str(fav))
    return dot


visualize_leetcode_maximum_employees_to_be_invited_to_a_meeting([6,10,10,0,6,0,4,4,1,3,3])
        
````
